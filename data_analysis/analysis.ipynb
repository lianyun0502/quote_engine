```python
import pandas as pd
import numpy as np
import os
import json
from  typing import Dict

df = pd.DataFrame(columns=['timestamp', 'symbol', 'best_bid', 'best_bid_vol', 'best_ask', 'best_ask_vol', 'spread'])

f = open("../build/multi_bybit_spot/data/orderbook_202410230700.data")

for line in f:
    try:
        data = json.loads(line)
    except:
        print(line)
        continue
    if data['S'] != 'TIAUSDT' :
        continue
    Bids:Dict[str, str] = data['Bids']
    Asks:Dict[str, str] = data['Asks']
    bid_keys = list(Bids.keys())
    ask_keys = list(Asks.keys())
    best_bid, best_bid_vol = bid_keys[-1], Bids[bid_keys[-1]]
    best_ask, best_ask_vol = ask_keys[0], Asks[ask_keys[0]]
    df2 = pd.DataFrame({
        'timestamp':data['T'], 
        'symbol':data["S"], 
        'best_bid': best_bid, 
        'best_bid_vol':best_bid_vol, 
        'best_ask':best_ask, 
        'best_ask_vol':best_ask_vol
        }, index=[0])
    df = pd.concat([df, df2], ignore_index=True)
    continue
df
```
